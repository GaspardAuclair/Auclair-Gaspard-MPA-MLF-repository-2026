## Exercise 1.3

Write a script that will convert an arbitrary integer to a binary number (the number will be represented as a string with only *0* and *1*). Avoid using the *bin()* function or any other pythons default functions

```python
n = int(input("Enter an integer: "))

if n == 0:
    print("0")
else:
    negative = False
    if n < 0:
        negative = True
        n = -n

    binary = ""

    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2

    if negative:
        binary = "-" + binary

    print(binary)
```

## Exercise 1.4 - The Fibonacci Sequence:

The Fibonacci Sequence is a series of numbers. The following number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The following number in this series above is 8 + 13 = 21


Your task is to implement a function *fibonacci* that takes an integer as an input and returns a list that contains all Fibonacci numbers with values lower than the input integer

Example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonacci(upper_threshold: int) -> list:
    
    if upper_threshold <= 0:
        return []

    seq = [0]
    a, b = 0, 1
    while b < upper_threshold:
        seq.append(b)
        a, b = b, a + b
    return seq
```

Try to implement the function using multiple approaches: *Iterative approach*, *Recusrion*, *Memoization*

```python
def fibonacci_recursive(upper_threshold: int) -> list:
    def fib_sequence(a, b):
        if a >= upper_threshold:
            return []
        return [a] + fib_sequence(b, a + b)   # build list head-first

    return fib_sequence(0, 1)

print(fibonacci_recursive(10))
```

```python
def fibonacci_memoized(upper_threshold: int) -> list:
    cache = {}

    def fib(n):
        """Return the nth Fibonacci number (0-indexed)."""
        if n in cache:
            return cache[n]
        cache[n] = fib(n - 1) + fib(n - 2) if n > 1 else n
        return cache[n]

    result, n = [], 0
    while (value := fib(n)) < upper_threshold:
        result.append(value)
        n += 1
    return result

print(fibonacci_memoized(10))
```

## Exercise 1.5. -  Rock, Paper, Scissors game:

#### basic

*Rock, Paper, Scissors* is a well-known and straightforward game. If you do not know the rules, google them. We will write a code for the Rock, Paper, Scissors game where the user plays against a random computer. The code can be written into one function *rock_paper_scissors*.

Notes:
- This implementation uses the **random** library to enable the computer to make a random choice.
- Keyword **Input** may be helpful. Check it
- After the function is run, the program will ask you about your movement, which you type in
- The game is case-insensitive for user input (e.g., "rock", "Rock", and "ROCK" are all valid).
- This script plays one round of the game.
- The script prints the result ('You lose', 'You win', 'It is a tie')


```python
import random

def rock_paper_scissors(n: int) -> None:
    # Initialize scores and round counter
    user_score = 0
    computer_score = 0
    current_round = 0

    choices = ['rock', 'paper', 'scissors']

    print(f"--- Starting a {n} round game! ---")

    while current_round < n:
        # 1. user input + normalize
        user_choice = input(f"\nRound {current_round + 1}: Choose rock, paper, or scissors: ").lower()

        if user_choice not in choices:
            print("Invalid choice! Please try again.")
            continue

        # 2. Computer choice
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        # 3. comparaison
        if user_choice == computer_choice:
            print("It's a tie for this round!")
        elif (user_choice == 'rock' and computer_choice == 'scissors') or \
             (user_choice == 'paper' and computer_choice == 'rock') or \
             (user_choice == 'scissors' and computer_choice == 'paper'):
            print("You win this round!")
            user_score += 1
        else:
            print("You lose this round!")
            computer_score += 1

        current_round += 1
        print(f"Score -> You: {user_score} | Computer: {computer_score}")

    print("\n--- Final Result ---")
    if user_score > computer_score:
        print(f"Final Score: {user_score}-{computer_score}. You win!")
    elif user_score < computer_score:
        print(f"Final Score: {user_score}-{computer_score}. You lose!")
    else:
        print(f"Final Score: {user_score}-{computer_score}. It is a tie!")

rock_paper_scissors(3)

```

## Exercise 2.1


In this exercise, you will work with NumPy arrays and learn how to efficiently modify and process them using both loops and vectorized operations. Your task is to:

1. Implement the function *create_array_nxn* that generates an n×n NumPy array filled with numbers from n² - 1 down to 0
2. Using a loop-based approach (for-loop, while-loop), implement the function *apply_threshold_loop* that takes any NumPy array and replaces all numbers below a user-defined threshold with 0.
3. Implement function *apply_threshold_vectorized* that performs the same operation as *apply_threshold_loop*, but without loops, using NumPy vectorized operations
4. Implement function *compare_performance* that compares the execution time of functions *apply_threshold_vectorized* and *apply_threshold_loop* with **time** library and print the results


**Additional Notes**:

- Ensure n is a positive integer (n > 0).
- Use the time library to measure execution times for performance comparison.
- The function *compare_performance(n, threshold)* should:
    - Generate an n×n array.
    - Apply both loop-based and vectorized thresholding.
    - Print execution times.

```python
import numpy as np
import time
```

```python
def create_array_nxn(n: int) -> np.ndarray:
    if not isinstance(n, int) or n <= 0:
        raise ValueError("n must be a positive integer (n > 0).")
    return np.arange(n ** 2 - 1, -1, -1).reshape(n, n)
```

```python
def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    result = arr.copy()
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            if result[i, j] < threshold:
                result[i, j] = 0
    return result
```

```python
def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
    result = arr.copy()
    result[result < threshold] = 0
    return result
```

```python
def compare_performance(n: int, threshold: int) -> None:
    arr = create_array_nxn(n)

    # Loop-based
    start = time.time()
    apply_threshold_loop(arr, threshold)
    loop_time = time.time() - start

    # Vectorized
    start = time.time()
    apply_threshold_vectorized(arr, threshold)
    vec_time = time.time() - start

    speedup = loop_time / vec_time if vec_time > 0 else float("inf")

    print(f"Array size      : {n}×{n}  |  Threshold: {threshold}")
    print(f"Loop time       : {loop_time:.6f} s")
    print(f"Vectorized time : {vec_time:.6f} s")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

GAP = np.zeros((5, 1), dtype=int)


def build_display_array(number: int) -> np.ndarray:
    digits = str(number)
    strips = []
    for i, ch in enumerate(digits):
        strips.append(numbs[ch])
        if i < len(digits) - 1:
            strips.append(GAP)
    return np.concatenate(strips, axis=1)


def show_digit(number: int) -> None:
    display = build_display_array(number)

    rows, cols = display.shape
    cell = 1
    pad  = 1.2

    fig_w = cols * cell + 2 * pad
    fig_h = rows * cell + 2 * pad

    fig, ax = plt.subplots(figsize=(max(fig_w * 0.6, 2), max(fig_h * 0.6, 2)))
    fig.patch.set_facecolor("black")
    ax.set_facecolor("black")

    bezel = patches.FancyBboxPatch(
        (-pad, -pad), cols * cell + 2 * pad - cell, rows * cell + 2 * pad - cell,
        boxstyle="round,pad=0.3",
        linewidth=2, edgecolor="#444444", facecolor="#111111",
        zorder=0
    )
    ax.add_patch(bezel)

    for r in range(rows):
        for c in range(cols):
            is_on = display[r, c] == 1
            y = (rows - 1 - r) * cell

            if is_on:
                rect = patches.Rectangle(
                    (c * cell, y), cell * 0.85, cell * 0.85,
                    linewidth=0, facecolor="white", zorder=2
                )
            else:
                rect = patches.Rectangle(
                    (c * cell, y), cell * 0.85, cell * 0.85,
                    linewidth=0, facecolor="#2a2a2a", zorder=1
                )
            ax.add_patch(rect)

    ax.set_xlim(-pad, cols * cell + pad - cell)
    ax.set_ylim(-pad, rows * cell + pad - cell)
    ax.set_aspect("equal")
    ax.axis("off")

    plt.show()
```

# Exercise 3 - Playing with California Housing Dataset

The purpose of this exercise is to learn basic operations of **pandas** library. Use *california_housing_dataset* you can find in google colab to do following operations:

- check what *dataset.describe()* does

- display all rows where *total_bedrooms* column is bigger than 310

- drop also first and last row

- save a mean of values in *households* columns to variable, using *matplotlib* display number of households in the graph (x-axis would be ID, y-axis - numbers of households) as dots with the same color, display also mean as a single line in the graph with different color as dots is

 - check if any of the columns contain NaN values, if yes, replace the NaN values with the arithmetic mean of the whole dataset

 - display *lat* and *long* into plot, *lat* would be on *x* axis and *long* on *y* axis

- choose two arbitrary columns and normalize all values inside using *min-max normalization*:

- Create a Correlation Matrix from all columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
raw = fetch_california_housing(as_frame=True)
df = raw.frame.rename(columns={"AveBedrms": "total_bedrooms", "Population": "households",
                                "Latitude": "lat", "Longitude": "long"})

# dataset.describe()
print("=== describe() ===")
print(df.describe())

# Rows where total_bedrooms > 310
filtered = df[df["total_bedrooms"] > 310]
print(f"\n=== Rows where total_bedrooms > 310 ({len(filtered)} rows) ===")
print(filtered)

# Drop first and last row
df = df.drop(index=[df.index[0], df.index[-1]]).reset_index(drop=True)
print(f"\n=== After dropping first & last row: {df.shape} ===")

# Mean of households + scatter plot with mean line
mean_households = df["households"].mean()
print(f"\n=== Mean households: {mean_households:.2f} ===")

plt.figure(figsize=(10, 4))
plt.scatter(df.index, df["households"], color="steelblue", s=5, label="Households")
plt.axhline(mean_households, color="red", linewidth=1.5, label=f"Mean = {mean_households:.1f}")
plt.xlabel("ID")
plt.ylabel("Number of households")
plt.title("Number of Households per District")
plt.legend()
plt.tight_layout()
plt.savefig("households_plot.png", dpi=120)
plt.show()

# Check & replace NaN with column mean
print("\n=== NaN counts per column ===")
print(df.isnull().sum())

for col in df.columns:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

print("NaN values replaced with column means.")

# Plot lat vs long
plt.figure(figsize=(7, 6))
plt.scatter(df["lat"], df["long"], s=1, color="darkgreen", alpha=0.4)
plt.xlabel("Latitude (lat)")
plt.ylabel("Longitude (long)")
plt.title("Geographic Distribution (lat / long)")
plt.tight_layout()
plt.savefig("lat_long_plot.png", dpi=120)
plt.show()

# Min-Max normalization on two columns
def min_max_normalize(series: pd.Series) -> pd.Series:
    return (series - series.min()) / (series.max() - series.min())

for col in ["MedInc", "HouseAge"]:
    df[f"{col}_normalized"] = min_max_normalize(df[col])

print("\n=== Min-Max normalization (MedInc, HouseAge) ===")
print(df[["MedInc", "MedInc_normalized", "HouseAge", "HouseAge_normalized"]].head(10))

# Correlation Matrix
corr = df.corr(numeric_only=True)
print("\n=== Correlation Matrix ===")
print(corr)

fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(corr, cmap="coolwarm", vmin=-1, vmax=1)
fig.colorbar(cax)
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=90, fontsize=8)
ax.set_yticklabels(corr.columns, fontsize=8)
plt.title("Correlation Matrix", pad=20)
plt.tight_layout()
plt.savefig("correlation_matrix.png", dpi=120)
plt.show()
```
